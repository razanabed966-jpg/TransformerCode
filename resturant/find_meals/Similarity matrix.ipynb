```python
import zipfile
import os

import pandas as pd

working_folder='/content/drive/My Drive/TransformersCode/03-restaurant/Find_meals/'

zip_file_path = working_folder +'indian_food.zip'

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(working_folder)

csv_file_path = os.path.join(working_folder, 'indian_food.csv')
df = pd.read_csv(csv_file_path)

df.head()
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/distiluse-base-multilingual-cased-v1')
```

```python
embeddings = model.encode(df['ARingredients'])
print(embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(embeddings, embeddings)
print(similarities.shape)
print(similarities)
```

```python
import numpy as np

similarities_file = working_folder + "cosine_similarity_matrix.npy"
np.save(similarities_file, similarities)
print(f"Similarity matrix saved to {similarities_file}")
```
