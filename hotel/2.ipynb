```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
```

```python
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    quantization_config=bnb_config
)
```

```python
from sentence_transformers import SentenceTransformer

ST = SentenceTransformer("distiluse-base-multilingual-cased-v1")
```

```python
working_folder='/content/drive/MyDrive/TransformersCode/04-hotel/'
```

```python
zip_file_path = working_folder + 'Hotel.zip'
import zipfile

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(working_folder)
```

```python
csv_file_path = working_folder + "Hotel.csv"
```

```python
import pandas as pd

df = pd.read_csv(csv_file_path)

df["text"] = df["Question"]+ " " + df["GoodAnswer"]

df["embeddings"] = df["text"].apply(lambda x: ST.encode(x))
df[["text","embeddings"]].head()
```

```python
from datasets import Dataset

dataset = Dataset.from_pandas(df)

dataset
```

```python
import faiss
import numpy as np


dataset = dataset.add_faiss_index(column='embeddings')
```

```python
SYS_PROMPT = """You are an assistant for answering questions.
 You are given the extracted parts of a long document and a question. Provide a conversational answer.
 Only give the answer Dont add any comments or explanations.
 If you don't know the answer, just say "I do not know." Don't make up an answer.
 Answer in Arabic if the question is in Arabic
 أجب فقط بالعربية إذا كان السؤال بالعربية
 """
```

```python
def search(question, k=2):

    embedded_question = ST.encode(question)

    scores, retrieved_documents = dataset.get_nearest_examples(
        "embeddings", embedded_question, k=k
    )
    return scores, retrieved_documents
```

```python
def format_prompt(question, retrieved_documents, k):
    PROMPT = f"Question: {question}\nContext:"
    for idx in range(k):
        PROMPT += f"{retrieved_documents['text'][idx]}\n"
    return PROMPT
```

```python
def talk(question):
    k=2
    scores, retrieved_documents = search(question, k)

    formatted_prompt = format_prompt(question, retrieved_documents, k)
    messages = [
        {"role": "system", "content": SYS_PROMPT},
        {"role": "user", "content": formatted_prompt}
    ]

    input_ids = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to("cuda")

    pad_token_id = tokenizer.pad_token_id if tokenizer.pad_token_id is not None else tokenizer.eos_token_id

    outputs_ids = model.generate(
        input_ids=input_ids,
        pad_token_id=pad_token_id,
        max_new_tokens=60,
        do_sample=True,
        temperature=0.01,
        top_p=0.95
    )

    output_text = tokenizer.decode(outputs_ids[0], skip_special_tokens=True)

    return output_text
```

```python
def get_last_line(s):
    lines = s.splitlines()
    return lines[-1] if lines else ''
```

```python
question = " أجب بالعربي هل يوجد مسبح في فندق الفورزد"

result = talk(question)
last_line = get_last_line(result)

print(last_line)
```
