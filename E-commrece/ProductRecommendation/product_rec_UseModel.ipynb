```python
import pandas as pd
import os

working_folder = '/content/drive/My Drive/TransformersCode/02-ECommerce/ProductRecommendations/'

csv_file_path = os.path.join(working_folder , 'flipkart_com-ecommerce_sample.csv')
df = pd.read_csv(csv_file_path)

df = df[['product_name']]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-mpnet-base-v2')
```

```python
import numpy as np

embeddings_path = working_folder + "embeddings.npy"

sentence_embeddings = np.load(embeddings_path)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

import re

def get_products(user_input, sentence_embeddings, model , df, num_recommendations=10):

  user_input = re.sub('[^A-Za-z0-9]+', ' ', user_input)

  user_input=user_input.lower()

  user_input_embedding = model.encode([user_input])

  similarities = cosine_similarity(user_input_embedding, sentence_embeddings)

  similarities=similarities[0]
  #print(similarities)

  indices_asc = similarities.argsort()

  indices_desc=indices_asc[::-1]

  top_indices=indices_desc[:num_recommendations]

  recommended_products = df.iloc[top_indices]['product_name']
  return recommended_products
```

```python
user_input="bluetooth wireless headphones"

recommended_products = get_products(user_input, sentence_embeddings, model, df)
print("Recommendations for \"",user_input,"\" are: ")
print(recommended_products)
```
